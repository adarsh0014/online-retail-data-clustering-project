## Data Cleaning

```python
cleaned_df = df.copy()
cleaned_df.info()
```

```python
# Convert the 'Invoice' column to a string data type
cleaned_df['Invoice'] = cleaned_df['Invoice'].astype(str)

# Create a mask to filter rows where 'Invoice' matches exactly 6 digits
mask = (cleaned_df['Invoice'].str.match('^\\d{6}$') == True)

# Apply the mask to filter the DataFrame and keep only matching rows
cleaned_df = cleaned_df[mask]

# Display the cleaned DataFrame
cleaned_df
```

```python
# Convert the 'StockCode' column to a string data type
cleaned_df['StockCode'] = cleaned_df['StockCode'].astype(str)

# Create a mask to filter rows where 'StockCode' matches:
# - Exactly 5 digits
# - 5 digits followed by letters
# - The exact string "PADS"
mask = (
    (cleaned_df['StockCode'].str.match('^\\d{5}$') == True) |
    (cleaned_df['StockCode'].str.match('^\\d{5}[a-zA-Z]+$') == True) |
    (cleaned_df["StockCode"].str.match("^PADS$") == True)
)

# Apply the mask to filter the DataFrame and keep only matching rows
cleaned_df = cleaned_df[mask]

# Display the cleaned DataFrame
cleaned_df
```

```python
# Drop rows where the 'Customer ID' column has missing values (NaN)
cleaned_df.dropna(subset=["Customer ID"], inplace=True)

# Display the cleaned DataFrame
cleaned_df
```

# Feature Engineering

```python
# Create a new column 'SalesLineTotal' by multiplying 'Quantity' and 'Price'
cleaned_df['SalesLineTotal'] = cleaned_df['Quantity'] * cleaned_df['Price']

# Display the updated DataFrame with the new column
cleaned_df
```

```python
# Group the cleaned DataFrame by 'Customer ID' and aggregate data
aggregated_df = cleaned_df.groupby(by='Customer ID', as_index=False).agg(
    MonetaryValue=("SalesLineTotal", 'sum'),  # Total sales per customer
    Frequency=('Invoice', 'nunique'),        # Number of unique invoices per customer
    LastInvoiceDate=('InvoiceDate', 'max')  # Most recent invoice date per customer
)

# Display the first few rows of the aggregated DataFrame
aggregated_df.head()
```

```python
# Find the maximum invoice date (most recent date across all customers)
max_invoice_date = aggregated_df['LastInvoiceDate'].max()

# Calculate the recency for each customer
# Recency is the number of days since the customer's last invoice date
aggregated_df['Recency'] = (max_invoice_date - aggregated_df['LastInvoiceDate']).dt.days

# Display the first 5 rows of the updated aggregated DataFrame
aggregated_df.head(5)
```

```python
import matplotlib.pyplot as plt

# Create a figure with 3 subplots to visualize Monetary Value, Frequency, and Recency distributions
plt.figure(figsize=(15, 5))

# Plot Monetary Value distribution
plt.subplot(1, 3, 1)
plt.hist(aggregated_df['MonetaryValue'], bins=10, color="skyblue", edgecolor="black")
plt.title("Monetary Value Distribution")
plt.xlabel("Monetary Value")
plt.ylabel("Counts")

# Plot Frequency distribution
plt.subplot(1, 3, 2)
plt.hist(aggregated_df['Frequency'], bins=10, color='lightgreen', edgecolor='black')
plt.title("Frequency Distribution")
plt.xlabel("Frequency")
plt.ylabel("Counts")

# Plot Recency distribution
plt.subplot(1, 3, 3)
plt.hist(aggregated_df['Recency'], bins=20, color='orange', edgecolor='black')
plt.title("Recency Distribution")
plt.xlabel("Recency")
plt.ylabel("Counts")

# Adjust layout for better spacing and display the plots
plt.tight_layout()
plt.show()
```

```python
# Calculate the first quartile (Q1) and third quartile (Q3) of the 'MonetaryValue' column
M_Q1 = aggregated_df['MonetaryValue'].quantile(0.25)  # 25th percentile
M_Q3 = aggregated_df['MonetaryValue'].quantile(0.75)  # 75th percentile

# Calculate the Interquartile Range (IQR)
IQR = M_Q3 - M_Q1

# Identify rows where 'MonetaryValue' is an outlier
# Outliers are values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
monetary_outlier_df = aggregated_df[
    (aggregated_df['MonetaryValue'] < (M_Q1 - 1.5 * IQR)) | 
    (aggregated_df['MonetaryValue'] > (M_Q3 + 1.5 * IQR))
].copy()

# Display descriptive statistics of the outlier DataFrame
monetary_outlier_df.describe()
```

```python
# Calculate the first quartile (Q1) and third quartile (Q3) of the 'Frequency' column
F_Q1 = aggregated_df['Frequency'].quantile(0.25)  # 25th percentile
F_Q3 = aggregated_df['Frequency'].quantile(0.75)  # 75th percentile

# Calculate the Interquartile Range (IQR)
IQR = F_Q3 - F_Q1

# Identify rows where 'Frequency' is an outlier
# Outliers are values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
frequency_outlier_df = aggregated_df[
    (aggregated_df['Frequency'] < (F_Q1 - 1.5 * IQR)) |
    (aggregated_df['Frequency'] > (F_Q3 + 1.5 * IQR))
]

# Display descriptive statistics of the outlier DataFrame
frequency_outlier_df.describe()
```

```python
# Filter the DataFrame to exclude rows that are outliers in 'MonetaryValue' and 'Frequency'
# Use `~` to negate the conditions, effectively selecting non-outlier rows
non_outlier_df = aggregated_df[
    (~aggregated_df.index.isin(monetary_outlier_df.index)) & 
    (~aggregated_df.index.isin(frequency_outlier_df.index))
]

# Display descriptive statistics for the non-outlier DataFrame
non_outlier_df.describe()
```

```python
# Create a figure with 3 subplots to visualize boxplots for non-outlier data
plt.figure(figsize=(15, 5))

# Boxplot for Monetary Value
plt.subplot(1, 3, 1)
sns.boxplot(data=non_outlier_df['MonetaryValue'], color='skyblue')
plt.title("Monetary Value Boxplot")
plt.xlabel("Monetary Value")

# Boxplot for Frequency
plt.subplot(1, 3, 2)
sns.boxplot(data=non_outlier_df['Frequency'], color='lightgreen')
plt.title("Frequency Boxplot")
plt.xlabel("Frequency")

# Boxplot for Recency
plt.subplot(1, 3, 3)
sns.boxplot(data=non_outlier_df['Recency'], color='orange')
plt.title("Recency Boxplot")
plt.xlabel("Recency")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

## K-means Clustering

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
# Define the maximum number of clusters to test
max_k = 12

# Initialize lists to store inertia and silhouette scores
inertia = []
silhoutte_scores = []

# Define the range of k values to test
k_values = range(2, max_k + 1)

# Loop through different values of k to calculate metrics
for k in k_values:
    # Initialize KMeans with current number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=1000)
    
    # Fit KMeans and predict cluster labels
    cluster_labels = kmeans.fit_predict(scaled_data_df)
    
    # Calculate silhouette score for current clustering
    sil_score = silhouette_score(scaled_data_df, cluster_labels)
    silhoutte_scores.append(sil_score)
    
    # Append the inertia (sum of squared distances to cluster centers)
    inertia.append(kmeans.inertia_)

# Plot the results
plt.figure(figsize=(15, 5))

# Inertia plot
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title("Line Plot of K Values vs Inertia")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)

# Silhouette score plot
plt.subplot(1, 2, 2)
plt.plot(k_values, silhoutte_scores, marker='o')
plt.title("Silhouette Scores for Different Values of k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.xticks(k_values)
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Initialize KMeans with 4 clusters
# - n_clusters=4: Number of clusters
# - random_state=42: Ensures reproducibility
# - max_iter=1000: Maximum number of iterations for the algorithm
kmeans = KMeans(n_clusters=4, random_state=42, max_iter=1000)

# Fit KMeans to the scaled data and predict cluster labels
cluster_labels = kmeans.fit_predict(scaled_data_df)

# Display the cluster labels for each data point
cluster_labels
```

```python
# Assign the cluster labels to the 'Cluster' column in the non-outlier DataFrame
non_outlier_df['Cluster'] = cluster_labels

# Display the updated DataFrame with cluster labels
non_outlier_df
```

```python
# Define colors for each cluster using a dictionary
cluster_color = {0: '#1f77b4',  # Blue
                 1: '#ff7f0e',  # Orange
                 2: '#2ca02c',  # Green
                 3: '#d62728'}  # Red

# Map the cluster labels to their respective colors
colors = non_outlier_df['Cluster'].map(cluster_color)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

# Plot the MonetaryValue, Frequency, and Recency with cluster colors
scatter = ax.scatter(non_outlier_df['MonetaryValue'], 
                     non_outlier_df['Frequency'], 
                     non_outlier_df['Recency'], 
                     c=colors,  # Apply cluster-specific colors
                     marker='o')

# Label the axes
ax.set_xlabel('Monetary Value')
ax.set_ylabel('Frequency')
ax.set_zlabel('Recency')

# Set the title for the plot
ax.set_title('3D Scatter Plot of Customer Data by Cluster')

# Display the plot
plt.show()
```

```python

# Create a figure with 3 subplots for violin plots
plt.figure(figsize=(12, 18))

# Violin plot for Monetary Value by Cluster
plt.subplot(3, 1, 1)
sns.violinplot(x=non_outlier_df['Cluster'], y=non_outlier_df['MonetaryValue'], palette=cluster_color, hue=non_outlier_df['Cluster'])
plt.title('Monetary Value by Cluster')
plt.ylabel('Monetary Value')

# Violin plot for Frequency by Cluster
plt.subplot(3, 1, 2)
sns.violinplot(x=non_outlier_df['Cluster'], y=non_outlier_df['Frequency'], palette=cluster_color, hue=non_outlier_df['Cluster'])
plt.title('Frequency by Cluster')
plt.ylabel('Frequency')

# Violin plot for Recency by Cluster
plt.subplot(3, 1, 3)
sns.violinplot(x=non_outlier_df['Cluster'], y=non_outlier_df['Recency'], palette=cluster_color, hue=non_outlier_df['Cluster'])
plt.title('Recency by Cluster')
plt.ylabel('Recency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Find the indices that are common to both monetary and frequency outliers
overlap_indices = monetary_outlier_df.index.intersection(frequency_outlier_df.index)

# Identify monetary outliers that are not frequency outliers
monetary_only_outliers = monetary_outlier_df.drop(overlap_indices)

# Identify frequency outliers that are not monetary outliers
frequency_only_outliers = frequency_outlier_df.drop(overlap_indices)

# Identify outliers that are both monetary and frequency outliers
monetary_and_frequency_outliers = monetary_outlier_df.loc[overlap_indices]

# Assign unique cluster labels to differentiate outlier types
monetary_only_outliers["Cluster"] = -1
frequency_only_outliers["Cluster"] = -2
monetary_and_frequency_outliers["Cluster"] = -3

# Combine all outlier DataFrames into a single DataFrame
outlier_clusters_df = pd.concat([monetary_only_outliers, frequency_only_outliers, monetary_and_frequency_outliers])

# Display the DataFrame with labeled outlier clusters
outlier_clusters_df
```

```python
# Define custom colors for each outlier cluster
cluster_colors = {-1: '#9467bd',  # Purple for monetary-only outliers
                  -2: '#8c564b',  # Brown for frequency-only outliers
                  -3: '#e377c2'}  # Pink for combined monetary and frequency outliers

# Create a figure for violin plots
plt.figure(figsize=(12, 18))

# Plot Monetary Value by outlier cluster
plt.subplot(3, 1, 1)
sns.violinplot(x=outlier_clusters_df['Cluster'], 
               y=outlier_clusters_df['MonetaryValue'], 
               palette=cluster_colors,
               hue = outlier_clusters_df['Cluster'])
plt.title('Monetary Value by Cluster')
plt.ylabel('Monetary Value')

# Plot Frequency by outlier cluster
plt.subplot(3, 1, 2)
sns.violinplot(x=outlier_clusters_df['Cluster'], 
               y=outlier_clusters_df['Frequency'], 
               palette=cluster_colors,
               hue = outlier_clusters_df['Cluster'])
plt.title('Frequency by Cluster')
plt.ylabel('Frequency')

# Plot Recency by outlier cluster
plt.subplot(3, 1, 3)
sns.violinplot(x=outlier_clusters_df['Cluster'], 
               y=outlier_clusters_df['Recency'], 
               palette=cluster_colors,
               hue = outlier_clusters_df['Cluster'])
plt.title('Recency by Cluster')
plt.ylabel('Recency')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
